```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = 'https://www.monster.com/jobs/q-it-jobs?page=1&so=p.h.p'
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.get_text()) #Extracts all text from the documnets and removes any HTML tags
```

```python
soup.find_all("img") #Returns list of all <img> tags in the HTML document
```

```python
soup.title #Was able to ind title using soup
```
